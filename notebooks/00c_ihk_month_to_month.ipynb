```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Lomba/DATATHON DICODING/inflasiwatch/data/raw/google_trends/google_trends_Medan.xlsx")
```

```python
df["Date"] = pd.to_datetime(df["Date"])

ihk_cols = [
    "IHK kel 01",
    "IHK kel 03",
    "IHK Kel 06",
    "IHK Umum"
]

for col in ihk_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
    
    df[f"MtM_{col}"] = (df[col].pct_change() * 100).round(2)

print(df.head())
```

```python
df.to_excel("D:/Lomba/DATATHON DICODING/inflasiwatch/data/processed/ihk_month_to_month/Medan_MtM.xlsx", index=False)
```
