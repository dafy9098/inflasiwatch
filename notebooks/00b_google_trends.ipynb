```python
from pytrends.request import TrendReq
import pandas as pd
import time

# Koneksi ke Google Trends
pytrends = TrendReq(hl='id-ID', tz=420)

# Daftar keyword leading indicator
keywords = [
    "harga beras",
    "harga cabai",
    "harga telur",
    "harga daging ayam",
    "harga minyak goreng",
    "harga bawang",
    "tiket pesawat",
    "tiket kereta",
    "harga bensin"
]

# Menyimpan hasil tiap keyword
all_data = pd.DataFrame()

for keyword in keywords:
    pytrends.build_payload(
        kw_list=[keyword],
        timeframe='2020-01-01 2026-03-31',
        geo='ID-JB'
    )
    
    trend = pytrends.interest_over_time()
    
    if not trend.empty:
        trend = trend[[keyword]].reset_index()
        trend.columns = ['date', keyword]
        
        if all_data.empty:
            all_data = trend
        else:
            all_data = all_data.merge(trend, on='date', how='outer')
    
    time.sleep(2)  # menghindari rate limit

# Sort tanggal
all_data = all_data.sort_values('date')
```
